### Parte 1: Carga y Limpieza (Pandas y NumPy)

1. Importar librerias:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2. Leer al archivo y guardarlo en un DF

```python
dataset = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(dataset)
```

3. Impimir primeras 5 filas y mostrar los tipos de datosde cada columna

```python
print("Primeras 5 filas:")
print(df.head())
print("Tipos de datos de cada columna:")
print(df.dtypes)
```

4. Identificar si hay valores nulos en la columna HP y reemplazar esos valores con la mediana la misma columna usando "np" y "pd"


```python
nulos_antes = df['horsepower'].isnull().sum()
print(f"-Valores nulos encontrados: {nulos_antes}")
mediana = np.nanmedian(df['horsepower'])
df['horsepower'] = df['horsepower'].fillna(mediana)
print(f"-Valores nulos después de la limpieza: {df['horsepower'].isnull().sum()}")
print("Contenido reemplazado de la columna Horsepower:")
print(df['horsepower'].head(10))
```

5. Crear una nueva columna llamada "relación_peso_potencia" que que sea el resultado de dividir la columna weight entre HP


```python
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
print("Columna 'relacion_peso_potencia' creada.")
print(df[['weight', 'horsepower', 'relacion_peso_potencia']].head(10))
```

### Parte 2: Visualización de Datos (Matplotlib)

1. Histograma

```python
plt.hist(df['acceleration'], color='lightgreen', edgecolor='black')
plt.title('Distribución de la Aceleración')
plt.xlabel('Aceleración')
plt.ylabel('Cantidad de Vehículos')
plt.show()
```

2. Gráfico de dispersión

```python
plt.scatter(df['displacement'], df['horsepower'], color='orange', alpha=0.5)
plt.title('Desplazamiento vs Caballos de Fuerza')
plt.xlabel('Desplazamiento del Motor')
plt.ylabel('Caballos de Fuerza')
plt.show()
```

3. Gráfico de barras

```python
promedio_por_pais = df.groupby('origin')['mpg'].mean()
promedio_por_pais.plot(kind='bar', color='red')
plt.title('Promedio de MPG por País de Origen')
plt.xlabel('País de Origen')
plt.ylabel('Millas por Galón (MPG)')
plt.show()
```
